```r
# 自定义画图函数, 将多个图拼接在一张图上
multiplot <- function(..., plotlist=NULL, file, cols=1, layout=NULL) {
  library(grid)

  # Make a list from the ... arguments and plotlist
  plots <- c(list(...), plotlist)

  numPlots = length(plots)

  # If layout is NULL, then use 'cols' to determine layout
  if (is.null(layout)) {
    # Make the panel
    # ncol: Number of columns of plots
    # nrow: Number of rows needed, calculated from # of cols
    layout <- matrix(seq(1, cols * ceiling(numPlots/cols)),
                    ncol = cols, nrow = ceiling(numPlots/cols))
  }

 if (numPlots==1) {
    print(plots[[1]])

  } else {
    # Set up the page
    grid.newpage()
    pushViewport(viewport(layout = grid.layout(nrow(layout), ncol(layout))))

    # Make each plot, in the correct location
    for (i in 1:numPlots) {
      # Get the i,j matrix positions of the regions that contain this subplot
      matchidx <- as.data.frame(which(layout == i, arr.ind = TRUE))

      print(plots[[i]], vp = viewport(layout.pos.row = matchidx$row,
                                      layout.pos.col = matchidx$col))
    }
  }
}
```

## 多元回归分析

首先建立$y$与所有自变量的回归方程:
$$
y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \beta_3 x_3 + \beta_4 x_4 + \beta_5 x_5 + \beta_6 x_6 + \beta_7 x_7 + \beta_8 x_8 + \beta_9 x_9
$$

```r
mdl0 = lm(Systol ~ ., data)
mdl0
```

```r
# 对回归方程的F检验
summary(mdl0)
```

$y$和$x_1 \sim x_9$建立多元线性回归模型，结果如上

其中`Height`、`Chin`、`Forearm`、`Calf`、`Pulse`的F检验结果不显著，需要进行广义F检验来判断

$H_0: \beta_5 = \beta_6 = \beta_7 = \beta_8 = \beta_9 = 0$

$H_1: \beta_5 \sim \beta_9$ 中至少一个不为0

```r
# 简模型的回归
mdl1 = lm(Systol ~ Age + Years + Fraction + Weight, data)
summary(mdl1)
```

```r
# 全模型的误差
anova(mdl0)
```

```r
# 简模型的误差
anova(mdl1)
```

$$
SSE(F) = 2172.58\\
MSE(F) = 74.92\\
df_F = 29\\
SSE(R) = 2629.71\\
df_R = 34
$$

广义F统计量公式  
$$
F^*=\frac{\frac{\text{SSE(R)  - SSE(F)}}{\text{df_R-df_F}}}{\text{MSE(F)}}=\frac{\frac{2629.71-2172.58}{34-29}}{74.92}=1.22037
$$

$F^* \sim F(5,29)$

```r
# p值
1-pf(1.22037, 5, 29)
```

p = 0.325 > 0.05, 故接收原假设, 即$x_5 \sim x_9$对因变量的解释没有贡献, 可以从模型中删去
