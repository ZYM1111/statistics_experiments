## 检验原序列的平稳性

```r
y = ts(c(5589, 9983, 11083, 13217, 16131, 19288, 19376, 24605, 27421, 38109, 54410, 67219, 44988, 35261, 36418, 41786, 49100, 54951, 43089, 42095, 53120, 68132, 76471, 80873, 83111, 78772, 88955, 84066, 95309, 110119, 111893, 111279, 107673, 113495, 118784, 124074, 130709, 135635, 140653, 144948, 151489, 150681, 152893, 157627, 162794, 163216, 165982, 171024, 172149, 164309, 167554, 178581, 193189, 204956, 224248, 249017, 269296, 288224, 314237, 330354), start = 1949, frequency = 1)
plot(y)
```

```r
acf(y)
```

```r
pacf(y)
```

## 进行一阶差分

```r
n = length(y)
diff_y = y[2:n] - y[1:n-1]
diff_y
```

```r
ts_diff_y = ts(diff_y,frequency = 1)
plot(ts_diff_y)
```

```r
acf(ts_diff_y)
```

## 对1阶差分后的序列进行白噪声检验

```r
Box.test(ts_diff_y, type = "Ljung-Box", lag=6)
Box.test(ts_diff_y, type = "Ljung-Box", lag=12)
```

## 对差分后序列建立ARMA模型

```r
pacf(ts_diff_y)
```

```r
# 对原序列拟合ARIMA(1,1,0)
library(forecast)
fit = arima(y, order=c(1,1,0), include.mean = T)
fit
```

```r
accuracy(fit)
```

```r
qqnorm(fit$residuals)
qqline(fit$residuals)
Box.test(fit$residuals, type="Ljung-Box")
```

```r
forecast(fit, 5)
plot(forecast(fit, 5), xlab="Year", ylab="Volume")
```
