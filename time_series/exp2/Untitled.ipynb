```r
library(forecast)
fit = arima(y, order=c(1,0,0), include.mean = T)
fit
```

```r
accuracy(fit)
```

```r
qqnorm(fit$residuals)
qqline(fit$residuals)
Box.test(fit$residuals, type="Ljung-Box")
```

```r
forecast(fit, 5)
plot(forecast(fit, 5), xlab="Year", ylab="Production")
```
